```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Survey data_Inflight Satisfaction Score.csv')
```

```python
df['score'] = pd.to_numeric(df['score'], errors='coerce')
```

```python
df = df[df['score'] > 3]
```

```python
df.head()
```

```python
df.sort_values(by=['flight_number', 'scheduled_departure_date'], inplace=True)
```

```python
df.head()
```

```python
num_satisfied_late = df[df['arrival_delay_minutes'] < 0].shape[0]
total_satisfied = df.shape[0]
```

```python
percentage_satisfied_late = (num_satisfied_late / total_satisfied) * 100
```

```python
percentage_satisfied_late
```

```python

labels = ['Satisfied Customers with Late Arrival', 'Satisfied Customers with Early Arrival']
sizes = [percentage_satisfied_late, 100 - percentage_satisfied_late]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  


plt.title('Satisfied customers and arrival')


plt.show()
```
