```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Inflight Service_Inventory data.csv')
```

```python
df.head()
```

```python
product_grouped = df.groupby('entree_description').agg({
     
    'entree_code': 'first', 
    'consumed_entree_count': 'sum'
}).reset_index()
```

```python
product_grouped['rank'] = product_grouped['consumed_entree_count'].rank(ascending=False,method='min').astype(int)
```

```python
new_table = product_grouped[[ 'entree_description', 'entree_code', 'consumed_entree_count', 'rank']]
```

```python
new_table.head(5)
```

```python
sorted_table = new_table.sort_values(by='rank')
```

```python
sorted_table.head(10)
```

```python
sorted_table.tail(20)
```

```python
sorted_table.to_csv('sorted_table.csv', index=False)
```
